# CHECKING FOR OUTLIERS

```python
columns_per_row = 3

# Calculate the total number of columns and the required number of rows
num_columns = len(df_1.columns)
num_rows = (num_columns // columns_per_row) + (num_columns % columns_per_row > 0)

# Create a figure with subplots
fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(15, num_rows * 4))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each column in df_1 and create box plots
for i, column in enumerate(df_1.columns):
    sns.boxplot(x=df_1[column], ax=axes[i])  # Create a box plot for the current column
    axes[i].set_title(f'Box Plot of {column}')  # Set title for each subplot
    axes[i].grid(visible=True, axis='y')  # Enable grid lines for better readability

# Hide any unused subplots if the number of columns is not a perfect multiple
for j in range(i + 1, len(axes)):
    axes[j].axis('off')  # Turn off the axes that are not used

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()  # Display the combined box plots
```

```python
#  IMPUTING OUTLIERS WITH IQR
# Example DataFrame with 200 columns
df = pd.DataFrame(np.random.randn(1000, 200), columns=[f'var_{i}' for i in range(200)])

# Apply IQR outlier treatment across all columns
for column in df_1.columns:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_1[column].quantile(0.25)
    Q3 = df_1[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap values outside of bounds
    df_1[column] = np.where(df_1[column] > upper_bound, upper_bound, 
                          np.where(df_1[column] < lower_bound, lower_bound, df_1[column]))
```

# AFTER IMPUTING OUTLIER WITH IQR METHOD
